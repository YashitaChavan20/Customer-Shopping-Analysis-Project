```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password=""
)

cursor = conn.cursor()
cursor.execute("CREATE DATABASE IF NOT EXISTS retail_analytics")
conn.close()
```

```python
engine = create_engine(
    "mysql+pymysql://root:@localhost/retail_analytics"
)
```

```python
df = pd.read_csv('main_customer_shopping_behavior.csv')
```

```python
df.to_sql(name = 'customer_shopping_analysis', con = engine, if_exists = 'replace', index = False)
```
